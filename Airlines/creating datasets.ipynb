```python
import pandas as pd
import os
import glob
from pathlib import Path  
  
# œáœÅŒ∑œÉŒπŒºŒøœÄŒøŒπŒøœçŒºŒµ œÑŒø glob Œ≥ŒπŒ± ŒΩŒ± Œ¥ŒπŒ±Œ≤Œ¨œÉŒøœÖŒºŒµ œåŒªŒ± œÑŒ± csv Œ±œÅœáŒµŒØŒ±
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
airline_id = 1 
airline_data = []
ratings = pd.DataFrame([], columns=['User', 'Review', 'Rating', 'UserId', 'AirlineId'])

# Œ¥ŒπŒ±œÑœÅŒ≠œáŒøœÖŒºŒµ œåŒªŒ± œÑŒ± csv Œ±œÅœáŒµŒØŒ±
for f in csv_files:
      
    df = pd.read_csv(f)
      
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
    print('Airline Name:', f.split("\\")[-1][:-4])
    airline_name = f.split("\\")[-1][:-4]
    airline_name = airline_name.replace('_', ' ')
    # print the content
    print('Content:')
    display(df.head())
    print()
    
    # Œ∏Œ≠œÑŒøœÖŒºŒµ œÑŒø user id Œ±œÄœå 1 ŒºŒ≠œáœÅŒπ œÑŒøŒΩ Œ±œÅŒπŒ∏Œºœå œÑœâŒΩ ŒµŒ≥Œ≥œÅŒ±œÜœéŒΩ
    df['UserId'] = df.index + 1

    # Œ∏Œ≠œÑŒøœÖŒºŒµ œÑŒø restaurant id, œÑŒø ŒøœÄŒøŒØŒø Œ±œÖŒæŒ¨ŒΩŒµœÑŒ±Œπ Œ∫Œ±œÑŒ¨ 1 œÉŒµ Œ∫Œ¨Œ∏Œµ iteration
    df['AirlineId'] = airline_id
    
    
    # Œ∫œÅŒ±œÑŒ¨ŒºŒµ œÑŒ± ŒµœÉœÑŒπŒ±œÑœåœÅŒπŒø Œ≥ŒπŒ± œÑŒ∑ŒΩ Œ¥Œ∑ŒºŒπŒøœÖœÅŒ≥ŒØŒ± œÑŒøœÖ airlines.csv
    airline_data_point = [airline_id, airline_name]
    airline_data.append(airline_data_point)
    
    # Œ∫œÅŒ±œÑŒ¨ŒºŒµ œåŒªŒ± œÑŒ± ratings Œ≥ŒπŒ± œÑŒ∑ŒΩ Œ¥Œ∑ŒºŒπŒøœÖœÅŒ≥ŒØŒ± œÑŒøœÖ ratings.csv
    ratings = ratings.append(df, ignore_index = True)
    
    
    print('New Content:')
    display(df.head())
    print()
    
    airline_id += 1
    
    
```

```python
df_airlines = pd.DataFrame(airline_data, columns=['AirlineId', 'Name'])
```
