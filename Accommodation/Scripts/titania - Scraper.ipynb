```python
for index, container in enumerate(containers):
    
    user = container.find("a", {"class": "ui_header_link uyyBf"})
    review = container.find("q", {"class": "QewHA H4 _a"})
    rating = container.find("div", {"class": "Hlmiy F1"})
    
    rating_value = 0;
    if (str(rating.next) == "<span class=\"ui_bubble_rating bubble_0\"></span>"):
        rating_value = 0
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_10\"></span>"):
        rating_value = 1
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_20\"></span>"):
        rating_value = 2
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_30\"></span>"):
        rating_value = 3
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_40\"></span>"):
        rating_value = 4
    else :
        rating_value = 5
    
    if(review):
        print("User is :" + user.text.strip() + "\n")
        print("Review is :" + review.text.strip() + "\n")
        print("Rating is : ",rating_value)
        data_point = [user.text.strip(),review.text.strip(), rating_value]
        data.append(data_point)
        print("\n")
        print("\n")
        
    if(index == 9):
        break #since there are 10 contaners with reviews per page
```

```python
for i in range(10,1600,10):

    driver = webdriver.Chrome(ChromeDriverManager().install())
    url = "https://www.tripadvisor.com/Hotel_Review-g189400-d228891-Reviews-or" + str(i) + "-Titania_Hotel-Athens_Attica.html#REVIEWS"
    print("url is \n" + url) 
    driver.get(url)
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()

    
    containers = soup.find_all("div", {"class": "YibKl MC R2 Gi z Z BB pBbQr"})

    
    for index, container in enumerate(containers):
    
        user = container.find("a", {"class": "ui_header_link uyyBf"})
        review = container.find("q", {"class": "QewHA H4 _a"})
        rating = container.find("div", {"class": "Hlmiy F1"})

        rating_value = 0;
        if (str(rating.next) == "<span class=\"ui_bubble_rating bubble_0\"></span>"):
            rating_value = 0
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_10\"></span>"):
            rating_value = 1
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_20\"></span>"):
            rating_value = 2
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_30\"></span>"):
            rating_value = 3
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_40\"></span>"):
            rating_value = 4
        else :
            rating_value = 5

        if(review):
            print("User is :" + user.text.strip() + "\n")
            print("Review is :" + review.text.strip() + "\n")
            print("Rating is : ",rating_value)
            data_point = [user.text.strip(),review.text.strip(), rating_value]
            data.append(data_point)
            print("\n")
            print("\n")

        if(index == 9):
            break #since there are 10 contaners with reviews per page
```

```python
df = pd.DataFrame(data, columns=['User', 'Review','Rating'])
```
