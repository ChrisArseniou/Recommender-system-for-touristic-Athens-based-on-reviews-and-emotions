```python
for index, container in enumerate(containers):
    
    user = container.find_all("a", {"class": "BMQDV _F G- wSSLS SwZTJ FGwzt ukgoS"})
    review_header  = container.find("span", {"class": "yCeTE"})
    review_text = container.find("div", {"class": "biGQs _P pZUbB KxBGd"})
    
    if(review_header):
        print("User is :" + user[0].text.strip() + "\n")
        print("Review header is :" + review_header.text.strip() + "\n")
        print("Review text is :" + review_text.text.strip() + "\n")
        data_point = [user[0].text.strip(),review_header.text.strip(), review_text.text.strip()]
        data.append(data_point)
        print("\n")
        print("\n")
        
    if(index == 10):
        break #since there are 10 contaners with reviews per page but the first one is empty
```

```python
for i in range(10,800,10):

    driver = webdriver.Chrome(ChromeDriverManager().install())
    url = "https://www.tripadvisor.com/Attraction_Review-g189400-d242840-Reviews-or" + str(i) + "-Herod_Atticus_Odeon-Athens_Attica.html#REVIEWS"
    print("url is \n" + url) 
    driver.get(url)
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()

    
    containers = soup.find_all("div", {"class": "_c"})

    
    for index, container in enumerate(containers):
    
        user = container.find_all("a", {"class": "BMQDV _F G- wSSLS SwZTJ FGwzt ukgoS"})
        review_header  = container.find("span", {"class": "yCeTE"})
        review_text = container.find("div", {"class": "biGQs _P pZUbB KxBGd"})

        if(review_header):
            print("User is :" + user[0].text.strip() + "\n")
            print("Review header is :" + review_header.text.strip() + "\n")
            print("Review text is :" + review_text.text.strip() + "\n")
            data_point = [user[0].text.strip(),review_header.text.strip(), review_text.text.strip()]
            data.append(data_point)
            print("\n")
            print("\n")

        if(index == 10):
            break #since there are 10 contaners with reviews per page
```

```python
df = pd.DataFrame(data, columns=['User', 'Review Title', 'Review'])
```
