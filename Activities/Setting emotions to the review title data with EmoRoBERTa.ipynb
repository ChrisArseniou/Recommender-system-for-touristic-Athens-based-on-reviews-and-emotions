```python
from transformers import pipeline
```

```python
emotion = pipeline('sentiment-analysis', model='arpanghoshal/EmoRoBERTa')
```

```python
def get_emotion_label(text):
  text = str(text) # απαραίτητο όταν εξάγουμε το συναίσθημα από τον τίτλο
  #text = text[:250] # απαραίτητο όταν εξάγουμε το συναίσθημα από τον κείμενο
  print((text))
  print('\n')
  return(emotion(text)[0]['label'])
```

```python
hotels = pd.read_csv('/content/drive/MyDrive/ΕΙΔΙΚΟ ΘΕΜΑ/To Do Reviews/activities.csv')
ratings = pd.read_csv('/content/drive/MyDrive/ΕΙΔΙΚΟ ΘΕΜΑ/To Do Reviews/ratings.csv')
df = hotels.merge(ratings, how="left", on="ActivityId")
df.head()
```

```python
df['EmotionFromTitle'] = df['ReviewTitle'].apply(get_emotion_label)
```
