```python
hotels = pd.read_csv('/content/drive/MyDrive/ŒïŒôŒîŒôŒöŒü ŒòŒïŒúŒë/To Do Reviews/activities.csv')
ratings = pd.read_csv('/content/drive/MyDrive/ŒïŒôŒîŒôŒöŒü ŒòŒïŒúŒë/To Do Reviews/ratings.csv')
df = hotels.merge(ratings, how="left", on="ActivityId")
df.head()
```
