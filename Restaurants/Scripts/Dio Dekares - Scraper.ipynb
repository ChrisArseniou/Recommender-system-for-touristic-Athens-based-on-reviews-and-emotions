```python
for container in (containers):
    
    user = container.find("div", {"class": "info_text pointer_cursor"})
    review = container.find("div", {"class": "prw_rup prw_reviews_text_summary_hsx"})
    rating = container.find("div", {"class": "ui_column is-9"})
    rating_value = 0;
    if (str(rating.next) == "<span class=\"ui_bubble_rating bubble_0\"></span>"):
        rating_value = 0
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_10\"></span>"):
        rating_value = 1
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_20\"></span>"):
        rating_value = 2
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_30\"></span>"):
        rating_value = 3
    elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_40\"></span>"):
        rating_value = 4
    else :
        rating_value = 5
    
    if(review):
        print("User is :" + user.text.strip() + "\n")
        print("Review is :" + review.text.strip() + "\n")
        print("Rating is : ",rating_value)
        data_point = [user.text.strip(),review.text.strip(), rating_value]
        data.append(data_point)
        print("\n")
        print("\n")
```

```python
for i in range(10,500,10):

    driver = webdriver.Chrome(ChromeDriverManager().install())
    url = "https://www.tripadvisor.com.gr/Restaurant_Review-g189400-d6721199-Reviews-or" + str(i) + "-Dio_Dekares_i_Oka-Athens_Attica.html#REVIEWS"
    print("url is \n" + url) 
    driver.get(url)
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()

    containers = soup.find_all("div", {"class": "review-container"})
    
    for container in (containers):
    
        user = container.find("div", {"class": "info_text pointer_cursor"})
        review = container.find("div", {"class": "prw_rup prw_reviews_text_summary_hsx"})
        rating = container.find("div", {"class": "ui_column is-9"})
        rating_value = 0;
        if (str(rating.next) == "<span class=\"ui_bubble_rating bubble_0\"></span>"):
            rating_value = 0
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_10\"></span>"):
            rating_value = 1
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_20\"></span>"):
            rating_value = 2
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_30\"></span>"):
            rating_value = 3
        elif (str(rating.next) == "<span class=\"ui_bubble_rating bubble_40\"></span>"):
            rating_value = 4
        else :
            rating_value = 5

        if(review):
            print("User is :" + user.text.strip() + "\n")
            print("Review is :" + review.text.strip() + "\n")
            print("Rating is : ",rating_value)
            data_point = [user.text.strip(),review.text.strip(), rating_value]
            data.append(data_point)
            print("\n")
            print("\n")
    
```

```python
df = pd.DataFrame(data, columns=['User', 'Review','Rating'])
```
