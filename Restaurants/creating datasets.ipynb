```python
import pandas as pd
import os
import glob
from pathlib import Path  
  
# χρησιμοποιούμε το glob για να διαβάσουμε όλα τα csv αρχεία
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
restaurant_id = 1 
restaurant_data = []
ratings = pd.DataFrame([], columns=['User', 'Review', 'Rating', 'UserId', 'RestaurantId'])

# διατρέχουμε όλα τα csv αρχεία
for f in csv_files:
      
    df = pd.read_csv(f)
      
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
    print('Restaurant Name:', f.split("\\")[-1][:-4])
    restaurant_name = f.split("\\")[-1][:-4]
    # print the content
    print('Content:')
    display(df.head())
    print()
    
    # θέτουμε το user id από 1 μέχρι τον αριθμό των εγγραφών
    df['UserId'] = df.index + 1

    # θέτουμε το restaurant id, το οποίο αυξάνεται κατά 1 σε κάθε iteration
    df['RestaurantId'] = restaurant_id
    
    
    # κρατάμε τα εστιατόριο για την δημιουργία του restaurants.csv
    restaurant_data_point = [restaurant_id, restaurant_name]
    restaurant_data.append(restaurant_data_point)
    
    # κρατάμε όλα τα ratings για την δημιουργία του ratings.csv
    ratings = ratings.append(df, ignore_index = True)
    
    
    print('New Content:')
    display(df.head())
    print()
    
    restaurant_id += 1
    
    
```

```python
df_restaurants = pd.DataFrame(restaurant_data, columns=['RestaurantId', 'Name'])
```

```python
df_ratings = pd.DataFrame([], columns=[])
df_ratings['UserId'] = ratings['UserId'] 
df_ratings['RestaurantId'] = ratings['RestaurantId']
df_ratings['Review'] = ratings['Review'] 
df_ratings['Rating'] = ratings['Rating'] 
```

```python
df_ratings = df_ratings.sort_values(by='UserId', ignore_index=True)
```
