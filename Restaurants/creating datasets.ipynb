```python
import pandas as pd
import os
import glob
from pathlib import Path  
  
# Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Ï„Î¿ glob Î³Î¹Î± Î½Î± Î´Î¹Î±Î²Î¬ÏƒÎ¿Ï…Î¼Îµ ÏŒÎ»Î± Ï„Î± csv Î±ÏÏ‡ÎµÎ¯Î±
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
restaurant_id = 1 
restaurant_data = []
ratings = pd.DataFrame([], columns=['User', 'Review', 'Rating', 'UserId', 'RestaurantId'])

# Î´Î¹Î±Ï„ÏÎ­Ï‡Î¿Ï…Î¼Îµ ÏŒÎ»Î± Ï„Î± csv Î±ÏÏ‡ÎµÎ¯Î±
for f in csv_files:
      
    df = pd.read_csv(f)
      
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
    print('Restaurant Name:', f.split("\\")[-1][:-4])
    restaurant_name = f.split("\\")[-1][:-4]
    # print the content
    print('Content:')
    display(df.head())
    print()
    
    # Î¸Î­Ï„Î¿Ï…Î¼Îµ Ï„Î¿ user id Î±Ï€ÏŒ 1 Î¼Î­Ï‡ÏÎ¹ Ï„Î¿Î½ Î±ÏÎ¹Î¸Î¼ÏŒ Ï„Ï‰Î½ ÎµÎ³Î³ÏÎ±Ï†ÏÎ½
    df['UserId'] = df.index + 1

    # Î¸Î­Ï„Î¿Ï…Î¼Îµ Ï„Î¿ restaurant id, Ï„Î¿ Î¿Ï€Î¿Î¯Î¿ Î±Ï…Î¾Î¬Î½ÎµÏ„Î±Î¹ ÎºÎ±Ï„Î¬ 1 ÏƒÎµ ÎºÎ¬Î¸Îµ iteration
    df['RestaurantId'] = restaurant_id
    
    
    # ÎºÏÎ±Ï„Î¬Î¼Îµ Ï„Î± ÎµÏƒÏ„Î¹Î±Ï„ÏŒÏÎ¹Î¿ Î³Î¹Î± Ï„Î·Î½ Î´Î·Î¼Î¹Î¿Ï…ÏÎ³Î¯Î± Ï„Î¿Ï… restaurants.csv
    restaurant_data_point = [restaurant_id, restaurant_name]
    restaurant_data.append(restaurant_data_point)
    
    # ÎºÏÎ±Ï„Î¬Î¼Îµ ÏŒÎ»Î± Ï„Î± ratings Î³Î¹Î± Ï„Î·Î½ Î´Î·Î¼Î¹Î¿Ï…ÏÎ³Î¯Î± Ï„Î¿Ï… ratings.csv
    ratings = ratings.append(df, ignore_index = True)
    
    
    print('New Content:')
    display(df.head())
    print()
    
    restaurant_id += 1
    
    
```

```python
df_restaurants = pd.DataFrame(restaurant_data, columns=['RestaurantId', 'Name'])
```

```python
df_ratings = pd.DataFrame([], columns=[])
df_ratings['UserId'] = ratings['UserId'] 
df_ratings['RestaurantId'] = ratings['RestaurantId']
df_ratings['Review'] = ratings['Review'] 
df_ratings['Rating'] = ratings['Rating'] 
```
